```python

import pandas as pd
from textblob import TextBlob


data = pd.read_csv('data.csv')


print("Project Description:")
print("This project aims to analyze the sentiment of reviews using a Sentiment Analyzer program.")


print("\nDataset Description:")
print("The dataset used for this analysis is 'data.csv'. It contains information about product reviews, including the review text, ratings, and other details.")


print("\nFirst 5 rows of the dataset:")
print(data.head())


data['sentiment'] = data['reviews.text'].apply(lambda review: TextBlob(str(review)).sentiment.polarity if pd.notnull(review) else 0)


print("\nSentiment scores for each review:")
print(data[['reviews.text', 'sentiment']])


average_sentiment = data['sentiment'].mean()
print("\nAverage Sentiment Score:", average_sentiment)


positive_reviews = len(data[data['sentiment'] > 0])
neutral_reviews = len(data[data['sentiment'] == 0])
negative_reviews = len(data[data['sentiment'] < 0])

positive_percentage = (positive_reviews / len(data)) * 100
neutral_percentage = (neutral_reviews / len(data)) * 100
negative_percentage = (negative_reviews / len(data)) * 100

print("\nPercentage of Positive Reviews:", positive_percentage)
print("Percentage of Neutral Reviews:", neutral_percentage)
print("Percentage of Negative Reviews:", negative_percentage)


```
